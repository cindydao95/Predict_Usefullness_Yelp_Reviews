## 1.2 Select only opened-restaurant businesses

```python
rest_df = combined.loc[(combined['RestaurantGroup'] == 'True') & (combined['is_open'] == 1),:]
```

```python
rest_df.info(show_counts = True)
```

```python
rest_df.rename
```

```python
def get_2020_year(date):
    match = re.search(r'2020-\d{2}-\d{2}',date)
    if match:
        return True
    else:
        return False
```

```python
bools = rest_df['date'].apply(get_2020_year)
```

```python
rest_df_update = rest_df.loc[bools,['ReviewId', 'Business ID', 'user_id', 'useful_x', 'stars', 'date',
       'review_text', 'review_count', 'yelping_since',
       'average_stars', 'Name', 'City', 'State', 'Postal_code',
       'business_stars', 'review_counts']]
```

```python
rest_df_update=rest_df_update.rename(columns = {'useful_x':'useful','review_count':'user_review_counts','review_counts':'biz_review_counts'})
```
